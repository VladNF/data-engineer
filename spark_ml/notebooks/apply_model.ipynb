```scala
import org.apache.spark.sql.types.{StructType, StringType}
import org.apache.spark.ml.{Pipeline, PipelineModel}
import org.apache.spark.sql.functions._
import org.apache.spark.sql.DataFrame
```

```scala
println(s"Current spark version is ${spark.version}")
```

```scala
val inputStreamPath = "/home/jovyan/data/events-stream"
val modelPath = "/home/jovyan/models/spark-ml-model"

val dataSchema = new StructType()
    .add("tweet", StringType)

val inputDF = spark
    .readStream
    .schema(dataSchema)
    .option("maxFilesPerTrigger", 1)
    .json(inputStreamPath)
```

```scala
val predictionsDF: DataFrame = inputDF
```

```scala
val classifier = PipelineModel.load("/home/jovyan/models/spark-ml-model")
```

```scala
import org.apache.spark.sql.functions._
val getProbability = udf((prediction: org.apache.spark.ml.linalg.Vector) => prediction(1))
```

```scala
predictionsDF.writeStream.foreachBatch { (batchDF: DataFrame, batchId: Long) =>
    classifier
    .transform(batchDF)
    .select('tweet, 'prediction, getProbability('probability).alias("positive_probability"))
    .show
}.start()
```

```scala
spark.stop()
```
